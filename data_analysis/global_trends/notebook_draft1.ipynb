```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.__version__
```

```python
## Data import and cleaning
```

```python
global_trends=pd.read_csv("kaggle2019/1- mental-illnesses-prevalence.csv")
```

```python
global_trends
```

```python
global_trends.head()
```

```python
global_trends.columns = global_trends.columns.str.lower().str.replace(' ', '_')
```

```python
global_trends.columns
```

```python
global_trends=global_trends.rename(columns = {'schizophrenia_disorders_(share_of_population)_-_sex:_both_-_age:_age-standardized':'schizophrenia',
       'depressive_disorders_(share_of_population)_-_sex:_both_-_age:_age-standardized':'depressive_disorders',
       'anxiety_disorders_(share_of_population)_-_sex:_both_-_age:_age-standardized':'anxiety_disorders',
       'bipolar_disorders_(share_of_population)_-_sex:_both_-_age:_age-standardized':'bipolar_disorders',
       'eating_disorders_(share_of_population)_-_sex:_both_-_age:_age-standardized':'eating_disorders'
           })
```

```python
global_trends.isna().sum()
```

```python
global_trends.info()
```

```python
global_trends.head()
```

```python
global_trends=global_trends.drop(columns='code',inplace=False)
```

```python
global_trends
```

```python
illneses=['schizophrenia','depressive_disorders','anxiety_disorders','bipolar_disorders','eating_disorders']
```

## Data Analysis

We stared basic analysis of what we have, making plots

```python
global_trends_by_years=global_trends.groupby('year')[illneses].mean()
```

```python
global_trends_by_years
```

```python
global_trends_by_years['schizophrenia'].plot() 
plt.axvline(x=2004, color='red', linestyle='--')   # Facebook started
plt.axvline(x=2016, color='green', linestyle='--')   # TikTok started )
```

```python
global_trends_by_years['depressive_disorders'].plot()
plt.axvline(x=2004, color='red', linestyle='--')   # Facebook started
plt.axvline(x=2016, color='green', linestyle='--')   # TikTok started )
```

```python
global_trends_by_years['anxiety_disorders'].plot()
plt.axvline(x=2004, color='red', linestyle='--')   # Facebook started
plt.axvline(x=2016, color='green', linestyle='--')   # TikTok started )
```

```python
global_trends_by_years['eating_disorders'].plot()

plt.axvline(x=2004, color='red', linestyle='--')   # Facebook started
plt.axvline(x=2016, color='green', linestyle='--')   # TikTok started )
```

```python
global_trends_by_years['total']=global_trends_by_years['schizophrenia']+\
global_trends_by_years['anxiety_disorders']+\
global_trends_by_years['bipolar_disorders']+\
global_trends_by_years['eating_disorders']+\
    global_trends_by_years['depressive_disorders']         
    
# add to make a total disordes column
```
